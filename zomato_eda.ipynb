```python
df_country=pd.read_excel('Zomatodataset/Country-Code.xlsx')
df_country
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df
```

```python
country_name=final_df.Country.value_counts().index
```

```python
country_value=final_df.Country.value_counts().values
```

```python
plt.pie(country_value[:3],labels=country_name[:3],autopct="%1.2f%%")
plt.title('Top 3 countries which uses Zomato')
```

Observation:- Zomato max records are from india the USA and UK

```python
final_df.columns
```

```python
rating=final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
rating
```

## Obeservation 
1. When rating is b/w 4.5-4.9 ----> excelent
2. When rating is b/w 4.0-4.4 ----> very good
3. When rating is b/w 3.5-3.9 ----> good
4. When rating is b/w 2.5-3.4 ----> average
5. When rating is b/w 1-2.4 ----> poor
6. When rating is 0 it means not rated


```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
```

```python
sns.barplot(x="Aggregate rating", y="Rating count",hue="Rating color",data=rating,palette=['white','red','orange','yellow','green','green'],width=3)
plt.title('Ratings')
```

### Observations:-
1. Max people have not voted 
2. if we ignore the 0 rating colums our plot resembles a gaussian curve

```python
sns.countplot(x='Rating color',data=rating,palette=['blue','red','orange','yellow','green','green'])
```
